```python
import json
import pandas as pd
import glob
from pymongo import MongoClient
```

```python
converted_data = []

# Membaca data dari file JSON
# D:\documents\polban\sem 5\SI\raw
for filename in glob.glob('D:/documents/polban/sem 5/SI/raw/*.json'):
  with open(filename, mode='r', encoding='utf8') as json_file:
      datas = json.loads(json_file.read())

  # Mendapatkan informasi dari item-item dalam array items
  for data in datas:
      item = data['item']
      label = filename.split('.')[0].split('\\')[1]
      url_video = f"https://www.tiktok.com/@{item['author']['uniqueId']}/video/{item['id']}"
      like = item['stats']['diggCount']
      comment = item['stats']['commentCount']
      views = item['stats']['playCount']
      share = item['stats']['shareCount']
      save = item['stats']['collectCount']
      tanggal = item['createTime']
      engagement_total = (like + comment + share + save)

      converted_data.append({
          'label': label,
          'url_video': url_video,
          'like': like,
          'comment': comment,
          'views': views,
          'share': share,
          'save': save,
          'tanggal': tanggal,
          'engagement_total': engagement_total
      })

      # Print data
      print("Label:", label)
      print("URL video:", url_video)
      print("Like:", like)
      print("Comment:", comment)
      print("Views:", views)
      print("Share:", share)
      print("Save:", save)
      print("Tanggal:", tanggal)
      print("Engagement Total:", engagement_total)
      print("\n")
```

```python
mongo_client = MongoClient('mongodb://localhost:27017/')
db = mongo_client['tiktokDB']
collection = db['tiktok']

# Memasukkan data ke MongoDB
collection.insert_many(converted_data)
```
